```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread('400_300.jpg')

x = np.arange(-3,3,0.1)

sigmoid_x=1/(1+np.exp(-x))

relu=[]
for i in x:
  if i > 0:
    relu.append(i)
  else:
    relu.append(0)

plt.subplot(2,2,1)
plt.plot(x, sigmoid_x)
plt.title("Sigmoid")

plt.subplot(2,2,2)
plt.plot(x,sigmoid_x, label='sigmoid')
plt.plot(x,relu, label='relu')
plt.legend()
plt.title("Relu & Sigmoid")

plt.subplot(2,2,3)
plt.imshow(img)

plt.subplot(2,2,4)
plt.imshow(img)


plt.show()
```
